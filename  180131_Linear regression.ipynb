```python
############
#그래프 구현
############

import tensorflow as tf

#Linear regression의 구조는 1차 방정식이다. "x 이걸 넣으면 y에는 이게 나와야 한다"는 걸 컴퓨터에게 먼저 알려줌.
#H(x) = Wx + b training data 입력! 
x_train = [1,2,3]
y_train = [1,2,3]
 
#placeholder 사용!
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

#Variable 을 사용할때 값을 정할 수 없으니 random을 넣어줌.
W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

# Hypothesis = Wx+b => Regression방법에서 학습을 통한 직선을 나타냄
hypothesis = X * W + b

#cost(W,b) costFunction => 차의 제곱의 평균을 찾는 것(tf.reduce_mean은 평균을 내주는 함수)
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

#학습률과 최적화함수를 'optimizer' 텐서에 입력시켜줌.
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)

#optimizer를 이용해 cost를 최소화(minimize)화 시킴 즉 학습시킨다!
#결국 텐서플로우는 노드들처럼 이어져있는 구조로 텐서 'train' 이 루트 노드가 된다
train = optimizer.minimize(cost)

############
#그래프 실행
############

#그래프를 실행하기위한 session함수 선언
sess = tf.Session()

# sess을 실행하기위해서는 그래프 구현에서 만든 텐서들을 초기화 시켜줘야하는 과정이 필요함
sess.run(tf.global_variables_initializer())

#텐서'train'이 모든 텐서의 루트 노드이므로 run시키게 되면 관련된 모든 텐서(노드)들이 전부 실행되게 된다.

######################
#학습시작(2001번학습!)
######################
for step in range(2001):
    #run을 시키고 싶은 텐서들을 []로 묵어서 한번에 시켜준다.
    #앞에 cost_val, W_val 등은 []안에 쓰인 순서대로 새로운 별명을 붙여줌
    #마지막에 '_'는 따로 쓸일이없어 별명을 안붙여준것임
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train],
                                         feed_dict={X: x_train, Y: y_train})
    
    if step % 20 == 0:
        #앞에서 붙인 별명으로 출력해줌.
        print(step, cost_val, W_val, b_val)
        
        
#####################
#학습된 그래프 테스트
#####################
print("X: 4, Y: ", sess.run(hypothesis, feed_dict={X: 4}))
print("X: 7.5, Y: ", sess.run(hypothesis, feed_dict={X: 7.5}))
```
